```python
import geopandas as gpd

gdf = gpd.read_file('fl_census_tract.json') # geojson file
pdf = pd.read_csv('fl_census_data_final.csv') # CSV file
pdf['GEOID'] = pdf['GEOID'].astype(str)

joined_gdf = gdf.merge(pdf, on='GEOID', how = 'left')
joined_gdf.to_file('fl_census_data_map.geojson', driver = 'GeoJSON')
```
